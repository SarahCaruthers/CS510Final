Let's look at a correlation plot to confirm this theory.

```r
install.packages("ggcorrplot")
```

```r
library(corrplot)
data <- select(stack,tensorflow, python, numpy, pandas, r, deep.learning, machine.learning)
correlation_matrix <- round(cor(data),1)
corrp.mat <- cor_pmat(data)
head(corrp.mat[, 1:4])
```

```r
ggcorrplot(correlation_matrix, method ="circle")
```

It seems that programmers do not use r very much with tensorflow and deep.learning. It seems like python, tensorflow, and pandas is used equally with machine.learnging and deep.learning. R and numpy is used more with machine.learning. I will now do a time-series analysis on machine.learning and deep.learning and use those results to make inferences about the future searches of r, python, numpy, pandas, and tensorflow.

```r
ml <- select(stack, machine.learning)

ml_ts <- ts(ml, start = 2009, end = 2019, frequency = 12)

plot(ml_ts, xlab ="Monthly counts",
        ylab ="StackOverflow Searches",
        main ="Time Series of Machine Learning StackOverflow Searches",
        col.main ="darkgreen")
```

```r
install.packages("TSA")
install.packages("forecast")
```

```r
library(forecast)
library(TSA)
arima_model = auto.arima(ml_ts)
```

```r
pred <- forecast(arima_model, h = 20)
```

```r
plot(forecast(arima_model, h=20), xlab ="Monthly counts",
        ylab ="StackOverflow Searches",
        main ="Time Series of Machine Learning StackOverflow Searches",
        col.main ="darkgreen")
```

Let's find the seasonality, or monthly fluctuations of stackoverflow searches.

```r
trend_ml = ma(ml_ts, order = 4, centre = T)

detrend_ml = ml_ts - trend_ml
```

```r
plot(as.ts(detrend_ml), ylab =" ", main ="Seasonality of Machine Learning StackOverflow Searches",
        col.main ="darkgreen")
```

```r
dl <- select(stack, deep.learning)

dl_ts <- ts(dl, start = 2009, end = 2019, frequency = 12)

plot(dl_ts, xlab ="Monthly counts",
        ylab ="StackOverflow Searches",
        main ="Time Series of Deep Learning StackOverflow Searches",
        col.main ="darkgreen")
```

```r
library(forecast)
library(TSA)
arima_model2 = auto.arima(dl_ts)
```

```r
pred2 <- forecast(arima_model2, h = 20)
```

```r
plot(forecast(arima_model2, h=20), xlab ="Monthly counts",
        ylab ="StackOverflow Searches",
        main ="Time Series of Deep Learning StackOverflow Searches",
        col.main ="darkgreen")
```

Let's find the seasonality, or monthly fluctuations of stackoverflow searches.

```r
trend_dl = ma(dl_ts, order = 4, centre = T)

detrend_dl = dl_ts - trend_dl
```

```r
plot(as.ts(detrend_dl), ylab =" ", main ="Seasonality of Deep Learning StackOverflow Searches",
        col.main ="darkgreen")
```
